##Pandas Series
A Pandas Series is a one-dimensional array of indexed data.

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
```

The contents can be accessed in the same way as for NumPy arrays, to the difference that when more than one value is selected, the type remains a Pandas Series.

```python
print(data[0],type(data[0]))
```

```python
print(data[2:],type(data[2:]))
```

The type Series wraps both a sequence of values and a sequence of indices, which we can access with the values and index attributes.


*   values are the contents of the series as a NumPy array
  



```python
print(data.values,type(data.values))
```


*   index are the indices of the serie




```python
print(data.index,type(data.index))
```

###Series Indices
The main difference between NumPy arrays and Pandas Series is the presence of this index field. By default, it is set (as in NumPy arrays) as 0,1,..,size_of_the_series but a Series index can be explicitly defined. The indices may be numbers but also strings. Then, the contents of the series have to be accessed using these defined indices.

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=['a', 'b', 'c', 'd'])
print(data)
```

```python
print(data['c'])
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=[1, 3, 4, 2])
print(data)
```

```python
print(data[2])
```

##Pandas DataFrames

DataFrames is a fundamental object of Pandas. Dataframes can be seen as an array of Series: to each index (corresponding to an individual for instance or a line in a table), a Dataframe maps multiples values; these values corresponds to the columns of the DataFrame which each have a name (as a string).


In the following example, we will construct a Dataframe from two Series with common indices.

```python
area = pd.Series( {'California': 423967, 'Texas': 695662, 'New York': 141297, 'Florida': 170312, 'Illinois': 149995})
population = pd.Series({'California': 38332521, 'Texas': 26448193, 'New York': 19651127, 'Florida': 19552860, 'Illinois': 12882135})
```

```python
states = pd.DataFrame({'Population': population, 'Area': area})
print(states,type(states))
```

In Jupyter notebooks, DataFrames are displayed in a fancier way when the name of the dataframe is typed (instead of using print)

```python
states
```

DataFrames have


*   index that are the defined indices as in Series

*   values that return a (2D) NumPy array with the contents







```python
print(states.index)
print(states.columns)
print(states.values,type(states.values),states.values.shape)
```

###Groupby

A grouped operation starts by specifying which groups of data that we would want to operate over. 

```python
tips_gb = tips.groupby(['sex', 'smoker'])
tips_gb
```

```python
tips_gb1 = tips.groupby(['sex'])['tip'].sum()
tips_gb1
```

03.08-split-apply-combine.png

```python
tips_gb2 = tips.groupby(['sex'])['tip'].mean()
tips_gb2
```

Groupby works by telling pandas a couple of columns. Pandas will look in your data and see every unique combination of the columns that you specify. Each unique combination is a group. So in this case we will have four groups: male smoker, female smoker, male non-smoker, female non-smoker.

The groupby object by itself is not super important.

Once we have these groups (specified in the groupby object), we can do three types of operations on it (with the most important being agg)

ABD7BC55-A024-46D0-A989-ADEE0F40E0AB.jpeg

###Aggregate

The aggregate operation aggregates all the data in these groups into one value. You use a dictionary to specify which values you'd like. For example look below, we are asking for both the mean and the min value of the tip column for each group:

```python
tips_agg = tips_gb.agg({
    'tip': ['mean', 'min', 'max'],
    'total_bill':'max' 
})

tips_agg
```

###Filter

The next common group operation is a filter. This one is pretty simple, we filter out member of groups that don't meet our criteria.

```python
from numpy.random.mtrand import randint
rng = np.random.RandomState(0)

df2 = pd.DataFrame({"key": ["A", "B", "C", "A","B", "C"], 
                    "data1": range(6),
                    "data2": rng.randint(0, 10, 6)},
                     columns = ["key", "data1", "data2"])

df2
```

```python
df2_gb = df2.groupby('key')
```

```python
for (key,group) in df2_gb:
  print(key)
  print(group)
  print()
```

Nous souhaitons garder seulement les regroupements qui ont mean pour data2 < 5

```python
def filter_fun(dataframe):
  return dataframe['data2'].mean() < 5
```

```python
from numpy.random.mtrand import randint
rng = np.random.RandomState(0)

a = pd.DataFrame({"key": ["A", "A"], 
                    "data1": range(2),
                    "data2": rng.randint(0, 10, 2)},
                     columns = ["key", "data1", "data2"])

a
```

```python
filter_fun(a)
```

```python
df2.groupby('key').filter(filter_fun)

```

##Combining DataFrames

###Concat

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': [3, 5], 'B': [1, 2]}, index = [0, 1])
df2 = pd.DataFrame({'A': [6, 7], 'B': [4, 9]}, index = [2, 3])
```

```python
df1
```

```python
df2
```

pour concaténer 2 dataframes (ou plus) ayant les mêmes colonnes les uns en dessous des autres :

```python
pd.concat([df1, df2])
```

on peut concaténer ainsi des dataframes même si les index sont identiques, ils seront juxtaposés avec des valeurs d'index répétées :

```python
df1 = pd.DataFrame({'A': [3, 5], 'B': [1, 2]})
df2 = pd.DataFrame({'A': [6, 7], 'B': [4, 9]})
pd.concat([df1, df2])
```

si les dataframes n'ont pas les mêmes colonnes, par défaut, des NaN sont mis aux endroits non définis : 

```python
df1 = pd.DataFrame({'A': [3, 5], 'B': [1, 2]})
df2 = pd.DataFrame({'A': [6, 7], 'C': [4, 9]})
pd.concat([df1, df2]) 
```

si les dataframes n'ont pas les mêmes colonnes et qu'on veut conserver seulement les colonnes communes, intersection (sans avoir de NaN) :

```python
pd.concat([df1, df2], join = 'inner')
```

on peut ignorer les valeurs de l'index dans la concaténation (met un index de 0 à n - 1) : 

```python
df1 = pd.DataFrame({'A': [3, 5], 'B': [1, 2]})
df2 = pd.DataFrame({'A': [6, 7], 'B': [4, 9]})
pd.concat([df1, df2], ignore_index = True) 
```

(le défaut de concaténation des lignes est axis = 0)
on peut concaténér des dataframes et des séries, par exemple 

```python
df1 = pd.DataFrame({'A': [3, 5], 'B': [1, 2]})
s = pd.Series([7, 9], name = 'C')
pd.concat([df1, s], axis = 1)
```

###Merge

Jointures : c'est le même principe qu'en sql :



*   jointure simple (inner) qui par défaut utilise les noms des colonnes qui sont communs :




```python
df1 = pd.DataFrame({'A': [3, 5], 'B': [1, 2]})
df2 = pd.DataFrame({'A': [5, 3, 7], 'C': [9, 2, 0]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on = ['A'])
```



*   jointure externe : les lignes qui n'ont pas la clef commune sont quand mêmes présentes (comme en sql) 




```python
 pd.merge(df1, df2, how = 'outer')
```



*   on peut aussi faire une jointure externe gauche ou droite comme en sql avec how = 'left' ou how = 'right'






*   on peut aussi faire une jointure sur des colonnes qui ne s'appellent pas pareil dans les 2 dataframes




```python
df1 = pd.DataFrame({'C': [3, 5], 'B': [1, 2]})
df2 = pd.DataFrame({'D': [5, 3, 7], 'E': [9, 2, 0]})
pd.merge(df1, df2, left_on = ['C'], right_on = ['D'])
```
