## Basic Plots: Line, Bar, and Scatter
In this section, we'll explore how to create three basic types of visualizations using Matplotlib: line plots, bar plots, and scatter plots. These visualizations are commonly used for exploring and communicating data.

### Creating a Line Plot
A line plot is a type of chart that displays data as a series of points connected by straight lines. It is commonly used to display trends over time or other continuous data. To create a line plot using Matplotlib, we can use the plot function. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

In this example, we first create an array of 100 evenly spaced points between 0 and 10 using the linspace function from NumPy. We then create a second array by taking the sine of each point in the first array. Finally, we use the plot function to create the line plot, and the show function to display the plot.

### Creating a Bar Plot
A bar plot is a type of chart that displays data as a series of rectangular bars. It is commonly used to compare the sizes of different categories. To create a bar plot using Matplotlib, we can use the bar function. Here's an example:

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]

plt.bar(x, y)
plt.show()
```

In this example, we create two arrays: one for the categories (x) and one for the heights of the bars (y). We then use the bar function to create the bar plot, and the show function to display the plot.



### Creating a Scatter Plot
A scatter plot is a type of chart that displays data as a collection of points. It is commonly used to visualize the relationship between two variables. To create a scatter plot using Matplotlib, we can use the scatter function. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes)
plt.show()
```

In this example, we create two arrays of 50 random numbers between 0 and 1 using the rand function from NumPy. We also create two additional arrays of random numbers to specify the colors and sizes of the points. Finally, we use the scatter function to create the scatter plot, passing in the x, y, c, and s arguments, and the show function to display the plot.

## Adding Labels and Titles to Plots
In this section, we'll explore how to add labels and titles to plots using Matplotlib. These elements are important for effectively communicating the information contained in your visualization.

### Adding a Title to a Plot
A title provides a brief description of the content of the plot. To add a title to a plot using Matplotlib, we can use the title function. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.show()
```

In this example, we use the title function to add a title to the plot. The argument of the title function is a string that specifies the text of the title.

### Adding Axis Labels
Axis labels provide information about the variables represented on the x and y axes. To add axis labels to a plot using Matplotlib, we can use the xlabel and ylabel functions. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

In this example, we use the xlabel and ylabel functions to add labels to the x and y axes, respectively. The arguments of the xlabel and ylabel functions are strings that specify the text of the labels.



### Adding Legends
A legend provides information about the different data series represented in the plot. To add a legend to a plot using Matplotlib, we can use the legend function. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sine Wave')
plt.plot(x, y2, label='Cosine Wave')
plt.legend()
plt.show()
```

## Multiple Plots on a Single Figure
In this section, we'll explore how to create multiple plots on a single figure using Matplotlib. This is useful when you want to display multiple related visualizations in one view, or when you want to compare different data series.

### Creating Subplots
To create subplots in Matplotlib, we can use the subplot function. The subplot function takes three arguments: the number of rows, the number of columns, and the index of the current subplot. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1)

axs[0].plot(x, y1)
axs[0].set_title('Sine Wave')
axs[1].plot(x, y2)
axs[1].set_title('Cosine Wave')

plt.show()
```

In this example, we use the subplots function to create a figure with two subplots: one with a sine wave and one with a cosine wave. We use the plot function to create each plot, and the set_title function to add a title to each subplot. We also use the axs variable to refer to the subplots.

## Customizing Subplots
We can customize subplots in various ways, such as changing the size and layout of each subplot, adding a colorbar, or sharing axes between subplots. Here's an example:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 2, figsize=(8, 6), sharex='col', sharey='row')

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Wave')
axs[0, 1].scatter(x, y2)
axs[0, 1].set_title('Cosine Wave')
axs[1, 0].hist(y1)
axs[1, 0].set_title('Sine Wave Histogram')
axs[1, 1].hist(y2)
axs[1, 1].set_title('Cosine Wave Histogram')

plt.colorbar(axs[0, 1].collections[0], ax=axs[:, 1], location='bottom')

plt.show()
```

In this example, we use the subplots function to create a figure with four subplots, arranged in a 2x2 grid. We use the figsize argument to set the size of the figure, and the sharex and sharey arguments to share the x and y axes between subplots.

We also use different types of plots, such as scatter plots and histograms, to demonstrate how to customize each subplot. Finally, we use the colorbar function to add a colorbar to the scatter plot in the second subplot.

## Saving plots to file
To save the plots created using Matplotlib, you can use the savefig() function provided by Matplotlib. The savefig() function allows you to save the plot in various file formats such as PNG, PDF, SVG, and more. Here are the steps to save a plot in Matplotlib:

1.   Create the plot you want to save using the Matplotlib library.
2.   Use the savefig() function to save the plot to a file.
Here's an example of how to save a plot in PNG format:


```python
import matplotlib.pyplot as plt

# create a plot
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
plt.plot(x, y)

# save the plot in PNG format
plt.savefig('myplot.png')
```

## Distribution plots - histogram, KDE, rug plot
Seaborn provides a number of built-in functions for creating distribution plots. Some of the most commonly used distribution plots include histograms, KDE plots, and rug plots.

Here's how you can create each of these types of plots in Seaborn:

### Creating a histogram
A histogram is a plot that shows the distribution of a dataset by dividing it into a set of bins and counting the number of observations that fall into each bin. You can create a histogram in Seaborn using the distplot() function. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create histogram
sns.distplot(tips["total_bill"], bins=10)

# Add labels and title
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.title("Histogram of Total Bill")
```

### Creating a KDE plot
A KDE (Kernel Density Estimation) plot is a smoothed version of a histogram that shows the density of a dataset. You can create a KDE plot in Seaborn using the kdeplot() function. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create KDE plot
sns.kdeplot(tips["total_bill"])

# Add labels and title
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.title("KDE Plot of Total Bill")
```

### Creating a rug plot 
A rug plot is a plot that shows the distribution of a dataset by placing a tick at the location of each observation along an axis. You can create a rug plot in Seaborn using the rugplot() function. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create rug plot
sns.rugplot(tips["total_bill"])

# Add labels and title
plt.xlabel("Total Bill")
plt.title("Rug Plot of Total Bill")
```

By combining these types of plots, you can create informative and visually appealing visualizations of your data in Seaborn.

### Creating a heat map
A heat map is a way to visualize two-dimensional data in a color-coded matrix, where each cell is assigned a color based on its value. You can create a heat map in Seaborn using the heatmap() function. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
flights = sns.load_dataset("flights")

# Create pivot table with the data
flights_pivot = flights.pivot("month", "year", "passengers")

# Create heat map
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Heat Map of Flight Passengers by Month and Year")
```
